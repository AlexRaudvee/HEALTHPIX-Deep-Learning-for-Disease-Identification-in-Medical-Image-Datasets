```python
from dc1.image_dataset import ImageDataset
from pathlib import Path
from image_rotate_resize import rotate_resize
from matplotlib import pyplot as plt
import torch
```

```python
train_dataset = ImageDataset(Path("../../data/X_train.npy"), Path("../../data/Y_train.npy"))
```

```python
images_dict = {}
x, y = train_dataset.imgs, train_dataset.targets
for x, y in zip(x, y):
    if int(y) in images_dict.keys():
        images_dict[int(y)].append(x)
    else:
        images_dict[int(y)] = [x]
```

```python
transformations_count = {}
for d_type in range(6):
    transformations_count[d_type] = {'':0, 'cropped':0, 'rotated':0, 'rotatedcropped':0}
    for n, image in enumerate(images_dict[d_type]):
        result_image, result_type = rotate_resize(image[0])
        transformations_count[d_type][result_type]+=1
        if result_type == 'rotatedcropped': #or cropped or rotated
            print(result_type)
            fig, arr = plt.subplots(ncols=2)
            arr[0].imshow(image[0], cmap='Greys', interpolation='nearest')
            arr[0].title.set_text('From')
            arr[1].imshow(result_image, cmap='Greys', interpolation='nearest')
            arr[1].title.set_text('To')
            plt.show()
```
